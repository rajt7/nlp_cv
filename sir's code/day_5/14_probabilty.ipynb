```python
import random

def coin_toss():
   return random.randint(0, 1)

tosses = [coin_toss() for _ in range(1000)]

prob_heads = tosses.count(0) / len(tosses)

print("Experimental probability of heads:", prob_heads)
```

```python
import random

def dice_roll():
  return random.randint(1, 6)

rolls = [dice_roll() for _ in range(1000)]

prob_six = rolls.count(6) / len(rolls)

print("Experimental probability of rolling a six:", prob_six)
```

### Markov Model

```python
import random

# Define states
states = ['Sunny', 'Rainy']

# Define transition matrix
transition_matrix = [
    [0.7, 0.3],
    [0.8, 0.2]
]

# Define Initial State
initial_state = 'Sunny'
```

```python
# Number of steps to simulate
num_steps = 10

# Generate Weather Sequence
weather_sequence = [initial_state]

for _ in range(num_steps):
    current_state = weather_sequence[-1]
    next_state_index = random.choices(
        range(len(states)),
        weights=transition_matrix[states.index(current_state)]
    )[0]
    next_state = states[next_state_index]
    weather_sequence.append(next_state)

# Print the generated weather sequence
print("Weather Sequence:", weather_sequence)
```
