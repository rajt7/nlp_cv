```python
# Find the synonyms of all the adjectives in the sentence

from nltk.tokenize import word_tokenize
from nltk import pos_tag

sent = '''The fiber is most often spun into yarn or thread and used to make a soft, rigid, and flexible textile.'''

tags = pos_tag(word_tokenize(sent))

# Getting the adjevtives.
adjectives = set()
for tag in tags:
    if tag[1].startswith('J'):
        adjectives.add(tag[0])

# Dictionary to store the adjectives along with their synonyms
adjective_dict = {}
for adj in adjectives:
    adjective_synonyms = []
    for synonym in wordnet.synsets(adj):
        if synonym.lemmas()[0].name() != adj:
            adjective_synonyms.append(synonym.lemmas()[0].name())
    adjective_dict[adj] = adjective_synonyms

adjective_dict
```

### Antonyms

```python
ant = set()
for synset in wordnet.synsets("run"):
    for lemma in synset.lemmas(): 
        if lemma.antonyms():             # When antonyms are available, add it.
            ant.add(lemma.antonyms()[0].name())

print("Antonyms: ", ant)
```
