```python
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
import numpy as np
from keras.utils import to_categorical
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
fig, ax = plt.subplots(5, 5)

# Flatten the axes array
ax = ax.flatten()

for i in range(25):
    ax[i].imshow(x_train[i], cmap='gray')
    ax[i].set_xticks([])
    ax[i].set_yticks([])
    ax[i].set_title(str(y_train[i]))

# Adjust the spacing
fig.tight_layout()

# Show the plot
plt.show()
```

```python
# Scaling 
x_train = x_train/255
x_test = x_test/255
```

```python
# reshaping 
x_train = x_train.reshape(-1,28,28,1)
x_test = x_test.reshape(-1,28,28,1)
```

```python
x_train.shape
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
# defining data preparation
datagen = ImageDataGenerator(zca_whitening=True, featurewise_center=True)
```

```python
datagen.fit(x_train)
```

```python
# configure batch size and retrieve one batch of images 
for x_batch, y_batch in datagen.flow(
    x_train, y_train, batch_size=9):

    # create a grd of 3x3 images
    for i in range(0, 9):
        plt.subplot(330+1+i)
        plt.imshow(x_batch[i].reshape(28, 28), cmap='gray')

    # show plot 
    plt.show()
    break
```

Random Flip

```python
# define data preparation
datagen = ImageDataGenerator(horizontal_flip=True, vertical_flip=True, featurewise_center=True)
```

```python
datagen.fit(x_train)
```

```python
# configure batch size and retrieve one batch of images
for X_batch, y_batch in datagen.flow(x_train, y_train, batch_size=9, shuffle=False):
    # create a grid of 3x3 images
    fig, ax = plt.subplots(3, 3, sharex=True, sharey=True, figsize=(4,4))
    for i in range(3):
        for j in range(3):
            ax[i][j].imshow(X_batch[i*3+j].reshape(28,28), cmap=plt.get_cmap("gray"))
    # show the plot
    plt.show()
    break
```

Zooming 

```python
datagen = ImageDataGenerator(zoom_range=0.5, featurewise_center=True)

datagen.fit(x_train)

for X_batch, y_batch in datagen.flow(x_train, y_train, batch_size=9, shuffle=False):
    # create a grid of 3x3 images
    fig, ax = plt.subplots(3, 3, sharex=True, sharey=True, figsize=(4,4))
    for i in range(3):
        for j in range(3):
            ax[i][j].imshow(X_batch[i*3+j].reshape(28,28), cmap="gray")
    # show the plot
    plt.show()
    break
```

Rotation

```python
datagen = ImageDataGenerator(rotation_range=60, featurewise_center=True)

datagen.fit(x_train)

for X_batch, y_batch in datagen.flow(x_train, y_train, batch_size=9, shuffle=False):
    # create a grid of 3x3 images
    fig, ax = plt.subplots(3, 3, sharex=True, sharey=True, figsize=(4,4))
    for i in range(3):
        for j in range(3):
            ax[i][j].imshow(X_batch[i*3+j].reshape(28,28), cmap="gray")
    # show the plot
    plt.show()
    break
```

Brightness

```python
datagen = ImageDataGenerator(brightness_range=(0.3, 4.5), featurewise_center=True)

datagen.fit(x_train)

for X_batch, y_batch in datagen.flow(x_train, y_train, batch_size=9, shuffle=False):
    # create a grid of 3x3 images
    fig, ax = plt.subplots(3, 3, sharex=True, sharey=True, figsize=(4,4))
    for i in range(3):
        for j in range(3):
            ax[i][j].imshow(X_batch[i*3+j].reshape(28,28), cmap="gray")
    # show the plot
    plt.show()
    break
```
