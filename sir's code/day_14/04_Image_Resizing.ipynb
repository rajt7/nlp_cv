```python
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
img = cv2.imread('../images/dog.jpg')
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### Resizing the Image

**Here, the aspect ratio is not changed and so the pixel values get blurry or high resoluted according to the resize.**

```python
# Resizing the Image
new_image = cv2.resize(img, (0,0), fx=1.75, fy=1.75)

print(f"Original Image shape: {img.shape}")
print(f"Resized Image shape: {new_image.shape}")

plt.figure(figsize=(16, 9))

plt.subplot(1,2,1)
# plt.grid()
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Resized Image')
plt.imshow(new_image);
```

```python
new_image = cv2.resize(img, (150, 200))

plt.figure(figsize=(16, 9))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Resized Image')
plt.imshow(new_image);
```
