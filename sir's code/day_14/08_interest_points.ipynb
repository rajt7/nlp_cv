```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
img = cv2.imread('../images/dog.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### Extract interest points

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
sift = cv2.SIFT_create()
key_points, _ = sift.detectAndCompute(gray, None)
new_img = cv2.drawKeypoints(gray, key_points, img, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(new_img)
```
