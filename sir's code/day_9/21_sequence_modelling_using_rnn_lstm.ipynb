```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
len(alphabets)
```

## Preprocess the data

```python
# Dictionary mapping from int to alphabets
int_to_char = {i: alphabet for i, alphabet in enumerate(alphabets)}

# Dictionary mapping from alphabets to int
char_to_int = {alphabet: i for i, alphabet in enumerate(alphabets)}
```

```python
# Creating a sequence length
seq_length = 1

# Creating the data
X = []
y = []

for i in range(len(alphabets) - seq_length):
    seq_in = [alphabets[i]]
    seq_out = alphabets[i+1]
    X += [[char_to_int[seq_in[0]]]]
    y += [char_to_int[seq_out]]
    # print(seq_in, '->', seq_out)
```

## One Hot encoding

```python
from keras.utils import to_categorical

y_new = to_categorical(y)
y_new
```

```python
y_new.shape
```

```python
import numpy as np

print(np.array(X).shape)
X = np.reshape(X, (25,1,1))
print(X.shape)
```

```python
X
```

## Scaling the data

```python
X = X / 25
X
```

## Build the Model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
model = Sequential()

model.add(Dense(25, input_shape=(1,1), activation='relu'))
model.add(LSTM(128))
model.add(Dense(26, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y_new, epochs=1000, batch_size=1)
```

```python
model.evaluate(X, y_new, batch_size=1)
```

## Predict on the new data
